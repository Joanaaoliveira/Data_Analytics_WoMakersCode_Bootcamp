### 4. De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos?
(Use o método ‘isin’, considere a média)

R: Para Advogado não, mas para Representante de Vendas sim

Considerando a qualidade do sono

```python
df_mediaqualidade_sono = df_saude[['Profissão','Qualidade do sono']].groupby('Profissão').mean().reset_index()
df_mediaqualidade_sono[df_mediaqualidade_sono['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]
```

Considerando a duração do sono

```python
df_mediaduracao_sono = df_saude[['Profissão','Duração do sono']].groupby('Profissão').mean().reset_index()
df_mediaduracao_sono[df_mediaduracao_sono['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]
```


### 8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

R: As mulheres têm maior pressão sanguínea


```python
df_presao_sanguinea = df_saude[['Gênero','Pressão sanguínea']]
```

```python
df_presao_sanguinea['pressao_sistolica'] = df_presao_sanguinea['Pressão sanguínea'].str.split('/', expand=True)[0].astype(int)
df_presao_sanguinea['pressao_diastolica'] = df_presao_sanguinea['Pressão sanguínea'].str.split('/', expand=True)[1].astype(int)
```

```python
df_presao_sanguinea.drop('Pressão sanguínea', axis=1, inplace=True)
```

```python
media_pressao = df_presao_sanguinea[['Gênero','pressao_sistolica', 'pressao_diastolica']].groupby('Gênero').mean()
media_pressao
```

### 10.  Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70?
(Use a média)


```python
media_passos = df_saude[['Frequência cardíaca', 'Passos diários']].groupby('Frequência cardíaca').mean().reset_index()
media_passos.head()
```

```python
passos_menor70 = media_passos[media_passos['Frequência cardíaca'] <= 70]['Passos diários'].sum()

passos_maior70 = media_passos[media_passos['Frequência cardíaca'] > 70]['Passos diários'].sum()

print(f'Pessoas com frequência cardíaca acima de 70 dão {passos_maior70:.2f} passos.\n'
      f'Pessoas com frequência cardíaca menor ou igual a 70 dão {passos_menor70:.2f} passos, considerando a média.\n'
      'Portanto, pessoas com frequência cardíaca acima de 70 dão mais passos que as abaixo de 70.')
```
